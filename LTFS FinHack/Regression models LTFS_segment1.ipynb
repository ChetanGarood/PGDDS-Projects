```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
```

```python
data11=pd.get_dummies(data1)
```

```python
data11.head()
```

```python
X=data11.drop(['application_date','Sum of case_count'],axis=1)
Y=data11[['Sum of case_count']]
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
reg=LinearRegression(normalize=True)   
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
from sklearn.metrics import r2_score,mean_squared_error
rmse=np.sqrt(mean_squared_error(y_pred,y_test))
r2=r2_score(y_test,y_pred)
print('rmse=',rmse)
print(r2_score(y_pred,y_test))
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor(loss='ls', n_estimators=80, max_depth=3, learning_rate=0.1, subsample=0.9)
xgb.fit(x_train,y_train)
y_predict=xgb.predict(x_test)
rmse=np.sqrt(mean_squared_error(y_predict,y_test))
print('rmse=',rmse)
print(r2_score(y_predict,y_test))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb=GradientBoostingRegressor(loss='ls', n_estimators=80, max_depth=3, learning_rate=0.1, subsample=0.9)
gb.fit(x_train,y_train)
y_predict=gb.predict(x_test)
rmse=np.sqrt(mean_squared_error(y_predict,y_test))
print('rmse=',rmse)
print(r2_score(y_predict,y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=155, max_depth=5)
rf.fit(x_train,y_train)
y_predict=rf.predict(x_test)
rmse=np.sqrt(mean_squared_error(y_predict,y_test))
print('rmse=',rmse)
print(r2_score(y_predict,y_test))
```
